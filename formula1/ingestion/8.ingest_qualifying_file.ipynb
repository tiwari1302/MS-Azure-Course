```
qualifying_df = spark.read.option("multiLine", True).option("inferSchema", True).json(f"{raw_folder_path}/qualifying")
```

#### Rename and add required columns

```
from pyspark.sql.functions import current_timestamp
```

```
final_df = qualifying_df.withColumnRenamed("driverId", "driver_id")\
.withColumnRenamed("raceId","race_id")\
.withColumnRenamed("constructorId","constructor_id")\
.withColumn("ingestion_date", current_timestamp())
```

#### Write data to processed containers in parquet format

```
final_df.write.mode("overwrite").parquet(f"{processed_folder_path}/qualifying")
```
