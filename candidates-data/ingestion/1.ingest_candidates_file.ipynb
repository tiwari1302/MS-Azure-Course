```
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType
```

```
candidates_schema = StructType(fields=[StructField("Application No", IntegerType(), False),
                                     StructField("Name", StringType(), True),
                                     StructField("Father Name", StringType(), True),
                                     StructField("DOB", DateType(), True)
                       ])
```

```
candidates_df = spark.read\
.option("header", True)\
.schema(candidates_schema)\
.csv(f"{raw_folder_path}/candidates.csv")\
.withColumnRenamed("Application No", "app_no")\
.withColumnRenamed("Name", "name")\
.withColumnRenamed("Father Name", "father_name")\
.withColumnRenamed("DOB", "dob")\
.dropDuplicates(['app_no'])
```

```
candidates_df.printSchema()
```
