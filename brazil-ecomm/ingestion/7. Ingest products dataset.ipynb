#### SELECTING REQUIRED COLUMNS FOR WRITING TO PROCESSED DIR

```
final_products_df = products_df.select('product_id', 'product_category_name')
```

```
final_products_df.write.mode('overwrite').parquet(f'{processed_folder_path}/products')
```
