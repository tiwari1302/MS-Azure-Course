```
from pyspark.sql.types import StringType, IntegerType, StructType, StructField
import optimus as op
```

```
schema = StructType([
            StructField("city", StringType(), True),
            StructField("country", StringType(), True),
            StructField("population", IntegerType(), True)])

countries = ['Colombia', 'US@A', 'Brazil', 'Spain']
cities = ['Bogotá', 'New York', '   São Paulo   ', '~Madrid']
population = [37800000,19795791,12341418,6489162]
```

```
# Dataframe:
df = op.spark.createDataFrame(list(zip(cities, countries, population)), schema=schema)
df.show()
```

```
# Instantiation of DataTransformer class:
transformer = op.DataFrameTransformer(df)

# Printing of original dataFrame:
print('Original dataFrame:')
transformer.show()

# Clear accents:
transformer.clear_accents(columns='*')

# Printing new dataFrame:
print('New dataFrame:')
transformer.show()
```

```
# Instantiation of DataTransformer class:
transformer = op.DataFrameTransformer(df)

# Printing of original dataFrame:
print('Original dataFrame:')
transformer.show()

# Remove special characters:
transformer.remove_special_chars(columns=['city', 'country'])

# Printing new dataFrame:
print('New dataFrame:')
transformer.show()

display(dbutils.fs.ls("/FileStore/tables/"))
```

```
# Import optimus
import optimus as op
# Instance of Utilities class
tools = op.Utilities()
# Reading df from web
df = tools.read_csv("/FileStore/tables/foo.csv", sep=",", header="true")
```

```
# Instantiation of DataTransformer class:
transformer = op.DataFrameTransformer(df)
# Replace NA with 0's
transformer.replace_na(0.0, columns="*")
# Show DF
transformer.show()
```

```
# Change Date formats  
# Instantiation of DataTransformer class:
transformer = op.DataFrameTransformer(df)

# Printing of original dataFrame:
print('Original dataFrame:')
transformer.show()

# Tranform string date format:
transformer.date_transform(columns="birth",
                          current_format="yyyy/mm/dd",
                          output_format="dd-MMM-yyyy")

# Printing new dataFrame:
print('New dataFrame:')
transformer.show()
```

```
# Change Column Names 

# Printing of original dataFrame:
print('Original dataFrame:')
transformer.show()

names = [('firstName', 'First Name')]
# Changing name of columns:
transformer.rename_col(names)

# Printing new dataFrame:
print('New dataFrame:')
transformer.show()
```

```
# replace strings 'piza', 'Cake' with pizza
transformer = op.DataFrameTransformer(df)

# Printing of original dataFrame:
print('Original dataFrame:')
transformer.show()

# Capital letters:
transformer.lookup('product', "pizza", ['piza', 'Cake'])

# Printing new dataFrame:
print('New dataFrame:')
transformer.show()
```

```
#move_col(column, ref_col, position)
# Instantiation of DataTransformer class:
transformer = op.DataFrameTransformer(df)

# Printing of original dataFrame:
print('Original dataFrame:')
transformer.show()

# Capital letters:
transformer.move_col('billingId', 'firstName', position='before')

# Printing new dataFrame:
print('New dataFrame:')
transformer.show()
```
